```
# replace $ with "" and M with 1000000 and B with 100000000
df["AMOUNT"] = df["AMOUNT"].str.replace("$", "").str.replace("M", "000000").str.replace("B", "000000000")
```

```
# remove . from the amount and make amount int type
df["AMOUNT"] = df['AMOUNT'].str.replace('.','').astype(int)
```

```
country_burrowers = df.groupby("Country")["BORROWER"].unique()
```

```
# build a new table with country, country_project_name, country_burrowers, country_lenders
country_detail = pd.DataFrame({
    # "Country" : df["Country"].unique(),
    "Project Name" : country_project_name,
    "Borrowers" : country_burrowers,
    "Lenders" : country_lenders
})
```
