```
import pandas as pd
```

```
df = pd.read_csv("top_insta_influencers_data.csv")
```

```
df.info()
```

```
df.isna().sum()
```

```
df["country"].fillna("unknown", inplace=True)
```

```
df["country"].value_counts()
```

```
df[["rank", "country"]].head(10).value_counts("country").plot(kind="bar")
```

```
df.sample(10)
```

```
df.head(10)
```

```
df["posts"] = df['posts'].str.replace("k",'').apply(lambda x:float(x)*1000 if "k" else float(x))
```

```
df.head()
```

```
df["followers"].sample(10)
```

```
df['followers'] = df['followers'].str.replace("m",'').apply(lambda x:float(x)*1000000 if "m" else float(x))
```

```
df.head()
```

```
df['avg_likes'] = df['avg_likes'].apply(lambda x: float(x.replace('k',''))*1000 if 'k' in x else (float(x.replace('m',''))*1000000 if 'm' in x else float(x)))
```

```
df.head()
```

```
df['new_post_avg_like'] = df['new_post_avg_like'].apply(lambda x: float(x.replace('k',''))*1000 if 'k' in x else (float(x.replace('m',''))*1000000 if 'm' in x else float(x)))
```

```
df.head()
```

```
df['total_likes'] = df['total_likes'].apply(lambda x: float(x.replace('k', ''))*1000 if 'k' in x else (float(x.replace('m', ''))*1000000 if 'm' in x else (float(x.replace('b', ''))*1000000000 if 'b' in x else float(x))))
```

```
# remove % from 60_day_eng_rate
df["60_day_eng_rate"] = df["60_day_eng_rate"].apply(lambda x: float(x.replace("%", "")))
```

```
corr_matrix = df[["posts", "followers", "60_day_eng_rate", "avg_likes", "new_post_avg_like", "total_likes"]].corr()
```

```
corr_matrix
```

```
# pairplot corr_matrix with sns

import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df[["posts", "followers", "60_day_eng_rate", "avg_likes", "new_post_avg_like", "total_likes"]])
plt.show()
```

```
# now show the corr_matrix in heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
