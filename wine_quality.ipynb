```
X = wine.drop(columns=['quality'], axis=1)
```

```
y = wine['quality']
```

```
y = wine['quality'].apply(lambda x:1 if x>= 7 else 0)
```

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```
random_model = RandomForestClassifier()
```

```
random_model.fit(X_train, y_train)
```

```
model = DecisionTreeClassifier()
```

```
# accuracy score for Random Forest Classifier

y_test_random_predicted = random_model.predict(X_test)
acc_test_random = accuracy_score(y_test_random_predicted, y_test)
acc_test_random
```

```
y_train_random_predicted = random_model.predict(X_train)
acc_train_random = accuracy_score(y_train_random_predicted, y_train)
acc_train_random
```

```
# Fit the DecisionTreeClassifier model
model.fit(X_train, y_train)

# Now you can predict using the trained model
y_test_logistic_predicted = model.predict(X_test)
acc_test_logistic = accuracy_score(y_test_logistic_predicted, y_test)
print(acc_test_logistic)
```

```
y_train_logistic_predicted = model.predict(X_train)
acc_train_logistic = accuracy_score(y_train, y_train_logistic_predicted)
acc_train_logistic
```
