```
expert_courses = df.loc[df["level"] == "Expert Level", ["course_title", "num_lectures","content_duration", "subject", "price","level", "url"]].sort_values("subject", ascending=False)
```

```
expert_courses
```

```
# plot the expert_courses by subject
expert_courses.value_counts("subject").plot(kind="bar", figsize=(10, 6), title="Expert courses by subject", x="subject", y="count")
```

```
expert_courses.loc[expert_courses["price"] > 100].plot(kind="bar", x="course_title", y=["num_lectures", "content_duration"])

```

```
expert_courses_correlation_matrix = expert_courses[["price", "num_lectures", "content_duration"]].corr()
```

```
expert_courses_correlation_matrix
```

```
# heatmap
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(expert_courses_correlation_matrix, annot=True)
plt.show()
```
