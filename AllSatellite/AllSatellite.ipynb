```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
pd.set_option('display.max_columns', 20)
```

```
df = pd.read_csv("AllSatellites.csv")
```

```
for item in df["Repeat Cycle (days)"].head(10):
  print(type(item))
```

```
df['Name'] = df['Name'].str.replace("\n", "")
df['Name'] = df['Name'].str.replace("\r", "")
```

```
df.head()
```

```
df.info()
```

```
df["Orbit Height (km)"] = df["Orbit Height (km)"].str.replace('\xa0', " ")
```

```
df["Repeat Cycle (days)"] = df["Repeat Cycle (days)"].str.replace('\xa0', " ")
```

```
df["Out of service since"] = df["Out of service since"].str.replace('\xa0', " ")
```

```
df["Launched in"] = df["Launched in"].str.replace('\xa0', " ")
```

```
df.info()
```

```
# replace " " with NaN and change type to float to "Orbit Height (km)" column

df["Orbit Height (km)"] = df["Orbit Height (km)"].replace(" ", np.nan).astype(float)
df["Repeat Cycle (days)"] = df["Repeat Cycle (days)"].replace(" ", np.nan).astype(float)
df["Launched in"] = df["Launched in"].replace(" ", np.nan).astype(float)
df["Out of service since"] = df["Out of service since"].replace(" ", np.nan).astype(float)
```
