```
df.groupby('author')['title'].count().sort_values(ascending=False).head(10)
# show mean usability_score of top contributor authors
top_usability_authors = df.groupby('author')['usability_score'].mean().sort_values(ascending=False).head(20)
top_usability_authors
```

```
# find mean usability for top_contributors
df[df['author'].isin(top_contributors.index)].groupby('author')['usability_score'].mean().sort_values(ascending=False)
```

```
df[df['author'].isin(top_contributors.index)].groupby('author')['usability_score'].mean().sort_values(ascending=False).plot(kind="bar", \
  figsize=(10, 5), xlabel="author", ylabel="mean usability", title="mean usability of top contributor authors")
```
